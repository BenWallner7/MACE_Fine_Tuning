## Split into training and testing set?

```python
graphene_water_protonic_defects_xyz
```

```python
def split_xyz(structures, train_size, test_size, validation_size, structures_included, seed):

    #Randomize the structures
    random.seed(seed)
    random.shuffle(structures)  

    # Calculate end point for validation structures
    
    val_end = train_size + validation_size

    # Split the data
    train_set = structures[:train_size]
    val_set = structures[train_size:val_end]
    test_set = structures[val_end:val_end + test_size]
    
    test_size = len(test_set)
    
    # Create folder

    folder = "/home/user/OneDrivePersonal/Documents/MPhil_Scientific_Computing/Written_Assignments/Fine_Tuning_MACE/training_datasets/{}_datasets_training_{}".format(structures_included, train_size)
    if folder and not os.path.exists(folder):  
        os.makedirs(folder)  
        
    write("/home/user/OneDrivePersonal/Documents/MPhil_Scientific_Computing/Written_Assignments/Fine_Tuning_MACE/training_datasets/{}_datasets_training_{}/train_{}_structures.xyz".format(structures_included, train_size, train_size), train_set)
    write("/home/user/OneDrivePersonal/Documents/MPhil_Scientific_Computing/Written_Assignments/Fine_Tuning_MACE/training_datasets/{}_datasets_training_{}/val_{}_structures.xyz".format(structures_included, train_size, validation_size), val_set)
    write("/home/user/OneDrivePersonal/Documents/MPhil_Scientific_Computing/Written_Assignments/Fine_Tuning_MACE/training_datasets/{}_datasets_training_{}/test_{}_structures.xyz".format(structures_included, train_size, test_size), test_set)

    print(f"Training set: {len(train_set)} structures")
    print(f"Validation set: {len(val_set)} structures")
    print(f"Testing set: {len(test_set)} structures")
```

## Two approaches for Fine Tuning

- Naive approach

```python
# Command line run

mace_run_train \
    --name="MACE" \
    --foundation_model="medium" \
    --multiheads_finetuning=False \
    --train_file="train.xyz" \
    --valid_fraction=0.05 \
    --test_file="test.xyz" \
    --energy_weight=1.0 \
    --forces_weight=1.0 \
    --E0s="average" \
    --lr=0.01 \
    --scaling="rms_forces_scaling" \
    --batch_size=2 \
    --max_num_epochs=6 \
    --ema \
    --ema_decay=0.99 \
    --amsgrad \
    --default_dtype="float64" \
    --device=cuda \
    --seed=3
```

- Multihead Relay approach

Prevents forgetting of previous training when fine-tuning


You need to select the --pt_train_file="mp" option

```python
mace_run_train \
    --name="MACE" \
    --foundation_model="medium" \
    --multiheads_finetuning=True \
    --train_file="train.xyz" \
    --valid_fraction=0.05 \
    --test_file="test.xyz" \
    --energy_weight=1.0 \
    --forces_weight=1.0 \
    --E0s="{"1": 130.0 }" \
    --lr=0.01 \
    --scaling="rms_forces_scaling" \
    --batch_size=2 \
    --max_num_epochs=6 \
    --ema \
    --ema_decay=0.99 \
    --amsgrad \
    --pt_train_file="mp" \
    --default_dtype="float64" \
    --device=cuda \
    --seed=3
```
